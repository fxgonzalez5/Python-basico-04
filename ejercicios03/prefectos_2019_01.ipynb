```python
import pandas as pd
import ipywidgets as widgets
```

```python
# generar el proceso para prefectos 
# leo y almaceno la data de los prefectos en una variable
datos_prefectos = pd.read_spss('data/resultadosdignidadprefecto2019.sav')
# Presento la data almacenada de prefectos
datos_prefectos
```

```python
# leo y almaceno la data de los candidatos en una variable
datos_candidatos = pd.read_spss('data/candidatos2019.sav')
# Presento la data almacenada de candidatos
datos_candidatos
```

```python
# Extraigo los campos que necesito trabajar de candidatoss en otra variable
datos_candidatos_2 = datos_candidatos[['CANDIDATO_CODIGO', 'CANDIDATO_NOMBRE', 'OP_CODIGO']]
# Presento la data extraída de los candidatos
datos_candidatos_2
```

```python
# Realizo unión entre los datos de los prefectos con los de candidatos y los almaceno en una variable
datosPrefectos = datos_prefectos.merge(datos_candidatos_2, on="CANDIDATO_CODIGO")
# Presento la data resultante luego de hacer la unión
datosPrefectos
```

```python
# leo y almaceno la data de las organizaciones políticas en una variable
organizaciones_politicas = pd.read_spss("data/organizacionespolticas2019.sav")
# Presento la data almacenada de organizaciones políticas
organizaciones_politicas
```

```python
# Extraigo los campos que necesito trabajar de las organizaciones políticas en otra variable
organizaciones_polticas_2 = organizaciones_politicas[['OP_CODIGO', 'OP_NOMBRE']]
# Presento la data extraída de las organizaciones política
organizaciones_polticas_2
```

```python
"""
Realizo unión entre los datos que contiene la vaiable "datosPrefectos" con los de organizaciones politicas
y sobeescribo la data en la misma variable
"""
datosPrefectos = datosPrefectos.merge(organizaciones_polticas_2, on="OP_CODIGO")
# Presento la data resultante luego de hacer la unión
datosPrefectos
```

```python
# Convierto la data almacenada en "datosPrefectos" en un archivo csv
datosPrefectos.to_csv("data/datos_prefectos_2019_modificado.csv", sep="|", index=False)
```

```python
# permitir buscar por provincias
# Leo y almaceno la data del csv anteriomente creado
datos_prefectosMod = pd.read_csv("data/datos_prefectos_2019_modificado.csv", sep="|")
# Presento la data extraida del csv
datos_prefectosMod
```

```python
# Creo una lista de tuplas con 2 elementos que contengan las provincias que estan dentro de la data
opcionesProvincia = [(i, i) for i in datos_prefectosMod.PROVINCIA_NOMBRE.unique()]
# Ordeno las provfincias en orden alfabético
opcionesProvincia.sort()
# Presento la lista de tuplas creada
opcionesProvincia
```

```python
# Creo una caja de opciones que contenga las provincias almacenadas anteriormente en la lista
provincia = widgets.Dropdown(options = opcionesProvincia, description = "Provincia:")
# Presento la caja de opciones
provincia
```

```python
# Realizo una consulta en la data extraída del csv cuando la povincia sea una en específica
datos_x_provincia = datos_prefectosMod[datos_prefectosMod['PROVINCIA_NOMBRE']==provincia.value]
# Presento la data tras hacer la consulta
datos_x_provincia
```

```python
# Creo un gráfico para mostrar estadisticamente la cantidad de votos de los prefectos de una provincia epecífica
datos_x_provincia.groupby("CANDIDATO_NOMBRE")['VOTOS'].sum().sort_values(ascending=False).plot.pie(ylabel='', figsize=(15,15))
```

```python
# Creo una lista de tuplas con 2 elementos que contengan los cantones que estan dentro de la data
opcionesCanton = [(i, i) for i in datos_prefectosMod.CANTON_NOMBRE.unique()]
# Ordeno las provfincias en orden alfabético
opcionesCanton.sort()
# Presento la lista de tuplas creada
opcionesCanton
```

```python
# Creo una caja de opciones que contenga los cantones almacenadas anteriormente en la lista
canton = widgets.Dropdown(options = opcionesCanton, description = "Cantón:")
# Presento la caja de opciones
canton
```

```python
# Realizo una consulta en la data extraída del csv cuando el cantón sea uno en específico
datos_x_canton = datos_prefectosMod[datos_prefectosMod['CANTON_NOMBRE']==canton.value]
# Presento la data tras hacer la consulta
datos_x_canton
```

```python
# Creo un gráfico para mostrar estadisticamente la cantidad de votos de los prefectos de un canton epecífico
datos_x_canton.groupby("CANDIDATO_NOMBRE")['VOTOS'].sum().sort_values().plot.barh(figsize=(10,10))
```
